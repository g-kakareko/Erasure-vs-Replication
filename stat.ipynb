```python
csv_fname = 'res_csv/rep_small.csv'
pd_rep_small = pd.read_csv(csv_fname) 

csv_fname = 'res_csv/rep_med.csv'
pd_rep_med = pd.read_csv(csv_fname)

csv_fname = 'res_csv/rep_large.csv'
pd_rep_large = pd.read_csv(csv_fname)

csv_fname = 'res_csv/jes_small.csv'
pd_jes_small = pd.read_csv(csv_fname)

csv_fname = 'res_csv/jes_med.csv'
pd_jes_med = pd.read_csv(csv_fname)

csv_fname = 'res_csv/jes_large.csv'
pd_jes_large = pd.read_csv(csv_fname)
```
