```python
# Figure 7- replication regression
t1 = pd_rep_small['time']
t25 = pd_rep_med['time']
t200 = pd_rep_large['time']

x_reg = np.concatenate([v_1, v_25, v_200], axis=0)
y_reg = np.concatenate([t1, t25, t200], axis=0)

slope, intercept, r_value, p_value, std_err = stats.linregress(x_reg, y_reg)
print(slope, intercept, r_value, p_value, std_err, r_value**2)
# x_p = np.array([1, 25,200])
x_p = np.arange(1,200)
y_p = slope*x_p + intercept 

z = np.polyfit(x_reg, y_reg, 2) 
p = np.poly1d(z)
y_pol = p(x_p)
# y = A + B log x
z_log = np.polyfit(np.log(x_reg), y_reg, 1)
y_log = z_log[0]*np.log(x_p) + z_log[1]
```

```python
plt.scatter(v_1, t1, edgecolor="k", s=80, color='blue')
plt.scatter(v_25, t25, edgecolor="k", s=80, color='green')
plt.scatter(v_200, t200, edgecolor="k", s=80, color='red')
plt.plot(x_p, y_p,'r--', linewidth=2.0)
plt.plot(x_p, y_pol,'b--', linewidth=2.0)
plt.plot(x_p, y_log,'g--', linewidth=2.0)

plt.legend(['lin reg','pol reg','log reg','small', 'medium', 'large'], loc='upper left')
plt.xlabel('Size [MB]')
plt.ylabel('Time [sec]')
# Tweak spacing to prevent clipping of ylabel
# axes = plt.gca()
# axes.set_xlim([0,0.2])
# axes.set_ylim([0,10])
fig = plt.gcf()
fig.set_size_inches(7.6, 7.6)
fig.savefig('figures/fig_7_rep_reg.png', dpi=300)
plt.show()
```

```python
# Figure 8- error replication regression
y_log = z_log[0]*np.log(x_reg) + z_log[1]
error = y_log - y_reg
plt.scatter(x_reg, error,marker='x', s=80, color='red')
plt.xlabel('Size [MB]')
plt.ylabel('error [sec]')
fig = plt.gcf()
fig.set_size_inches(7.6, 7.6)
fig.savefig('figures/fig_8_rep_reg_error.png', dpi=300)
plt.show()
```

```python
# Figure 9- q-q replication regression
i_vec = np.arange(1, len(x_reg)+1)
qi = np.array([(i-0.5)/len(i_vec) for i in i_vec])
zscore = norm.ppf(qi)

z = np.polyfit(zscore, np.sort(error), 1) 
p = np.poly1d(z)
y_pol = p(np.sort(zscore))

plt.plot(zscore, y_pol,'b--', linewidth=2.0)
# zscore
plt.scatter(zscore, np.sort(error),marker='x', s=80, color='red')
plt.xlabel('zscore')
plt.ylabel('error [sec]')
fig = plt.gcf()
fig.set_size_inches(7.6, 7.6)
fig.savefig('figures/fig_9_rep_qq.png', dpi=300)
plt.show()
```
